# Explanatory Data Analysis

## Gender analysis

```python
ax=sns.countplot(x='Gender',data=df)
# the variable ax refers to the matplotlib Axes object that is returned by the Seaborn countplot() function.
# We can also use any other variable name other than ax.
ax.set_title('Gender Distribution')
ax.set_xticklabels(['Male','Female'])
for bars in ax.containers:
    ax.bar_label(bars)

#ax.containers
#This returns a list of BarContainer objects.
#Each BarContainer holds one or more bars (individual rectangles).
#In a simple bar chart, you'll usually have just one container.

#for bars in ax.containers:
#You loop through each bar container (even if there's only one).

#ax.bar_label(bars)
#This function adds labels (text) to the bars in the container.By default, it labels the bars with their height, which corresponds to the count or value.
```

```python
sales_gen=df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
print(sales_gen)
```

```python
sns.barplot(x='Gender',y='Amount',data=sales_gen)
```

```python
# the above and this code does the same thing
sns.barplot(x='Gender',y='Amount',data=df,estimator=sum)
```

### Buying power per gender

```python
buying_power=(df.groupby('Gender')['Amount'].sum() / df.groupby('Gender')['User_ID'].nunique())
print(buying_power)
```

*From the above graph we can say female customer are more in number and have higher buying capacity*

## Age analysis

```python
ax=sns.countplot(x='Age Group',data=df,hue='Gender')
#Plot	Description
#sns.countplot(x='Age Group', data=df)	                One bar per age group (total count)
#sns.countplot(x='Age Group', data=df, hue='Gender')	Two bars per age group (count split by gender)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# Amount spend per age group
sales_age=df.groupby('Age Group',as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Age Group',y='Amount',data=sales_age)
```

*From the above graph we can see the age group residing between 26 to 35 are spending the most where most of them are female*

## State analysis

```python
# Orders from top 10 performing states
sales_state=df.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
sns.set(rc={'figure.figsize':(15,5)})
'''
sns.set() → Initializes Seaborn's theme and style settings.
rc={...} → Lets you pass matplotlib-style config options (runtime configuration = "rc").
'figure.figsize': (15, 5) → Sets the default figure size:
15 → width of the figure in inches
5 → height of the figure in inches
It makes all future plots wider and shorter — a landscape layout, useful when:
You have long category names on the x-axis (e.g., age groups)
You want to avoid label overlap
You're plotting grouped bars or detailed legends
'''
sns.barplot(x='State',y='Orders',data=sales_state)
```

```python
# total amount spent by top performing states
amt_state=df.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
sns.barplot(x='State',y='Amount',data=amt_state)
```

*Most of orders are made from UttarPradesh,Maharashtra and Karnataka.*

## Analysis based on Marital Status

```python
ax=sns.countplot(x='Marital_Status',data=df)
sns.set(rc={'figure.figsize':(7,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state = df.groupby(['Marital_Status', 'Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.set(rc={'figure.figsize':(6,5)})
sns.barplot(data = sales_state, x = 'Marital_Status',y= 'Amount', hue='Gender')
```

*From above graph we can see unmarried Female have more buying power*

```python
sales_occ=df.groupby(['Occupation'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(data=sales_occ,x='Occupation',y='Amount')
```

## Product Category Analysis

```python
ax=sns.countplot(x='Product_Category',data=df)
sns.set(rc={'figure.figsize':(15,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_cat=df.groupby(['Product_Category'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head()
sns.barplot(data=sales_cat,x='Product_Category',y='Amount')
sns.set(rc={'figure.figsize':(30,7)})
```

*From the above two graphs we can observe that Clothing,Food and Electronics and Gadgets are being sold the most*

```python
sales_prod=df.groupby(['Product_ID'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
sns.barplot(data=sales_prod,x='Product_ID',y='Orders')
sns.set(rc={'figure.figsize':(15,7)})
```
